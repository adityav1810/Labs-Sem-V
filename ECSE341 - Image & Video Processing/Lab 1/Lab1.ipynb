* ### Reading, saving, and displaying an image using
    - ##### PIL (Python Imaging Library). Print the width and height of the image 
    - ##### Matplotlib
    - ##### Skimage


```python
#PIL

from PIL import Image

#open
img = Image.open("data/apple.jpg")
img
```

```python
#width and height (size) along with some other info
print(img.format, img.size, img.mode)
```

```python
#save
img.save('data/out/apple.jpg')
```

```python
#check saved image
Image.open('data/out/apple.jpg')
```

```python
#matplotlib
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

imgmp = mpimg.imread('data/apple.jpg')
plt.imshow(imgmp)
```

```python
plt.imsave('data/out/mp_apple.jpg', imgmp)
```

```python
#checking saved file
plt.imshow(mpimg.imread('data/out/mp_apple.jpg'))
```

```python
#skimage
from skimage import io

sk_img = io.imread('data/apple.jpg')

io.imshow(sk_img)
#rest same as matplotlib
```

### Reading and displaying an image in Python without using special libraries such as OpenCV, PIL or Matplotlib. Write a program that reads a grayscale image in raw format from a file and save the new image into a new file in raw format. Numpy is the only standard library that can be used

```python
import numpy as np

file = open('data/bar.raw')
WIDTH = 512
HEIGHT = 512
img = np.fromfile(file, dtype = np.uint8, count = WIDTH*HEIGHT)
img.shape = (HEIGHT, WIDTH)

#Checking the created image using matplotlib
plt.imshow(img)
```

```python
#saving

img.astype('int8').tofile('data/out/img_created.raw')
```
