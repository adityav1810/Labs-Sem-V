```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
plt.imshow(rgb2gray(img1), cmap='gray')
```

#### Take two images and perform Image subtraction, addition, and multiplication of those images using arrays. 

```python
i_1 = Image.open('data/math_1.jpg')
i_2 = Image.open('data/math_2.jpg')
#making size equal
i_2 = i_1.resize((720,480))

i_1 = np.array(i_1)
i_2 = np.array(i_2)

#addition
i_a = i_1 + i_2
#subtraction
i_s = i_1 - i_2
#multiplication
i_m = i_1*i_2

plt.figure(1)
plt.subplot(311)
plt.imshow(i_a)
plt.subplot(312)
plt.imshow(i_s)
plt.subplot(313)
plt.imshow(i_m)


plt.imsave('data/math_a.jpg', i_a)
plt.imsave('data/math_s.jpg', i_s)
plt.imsave('data/math_m.jpg', i_m)
```
